## 1. Introduction

The aim of this coursework is building an image recognition system of P&ID images of the oil & gas industry. An initial repository of images was created with the images initially provided. The main processing procedures involved extracting the important features of these images through the methods of HOG and PCA. Another undersampling method used only for one dataset (agglomerative clustering decomposed PCA features dataset) was one sided selection over just the majority class.

The final two new repositories created were imbalanced, but with more images to each new category and with added variety in them.

Another main procedure was class decomposition. Specifically, two ways of class decomposition were followed, k-means clustering and hierarchical agglomerative clustering. For both the procedures, the existing classes of images were broken down to subclasses according to certain rules and algorithms so that indirectly the feature extracted datasets were better balanced. For the two decomposed datasets of the HOG feature data augmentation took place.

## 3. Feature extraction-Class decomposition

 

Two different image feature extraction algorithms were used: HOGs and PCA. 
The features extracted of each image in each case were then concatenated with the target/class 
the image initially belonged to and stored in a repository.
Each of these two repositories underwent:
    1. class decomposition 
    2. data augmentation  

After each class decomposition follows proof of the decomposition of the class through graphs.
Two methods of class decomposition are followed:
    1. K-means clustering
    2. Agglomerative clustering



Class Decomposition:
    
Kmeans clustering: It was used for class decomposition for features out of HOG algorithm and PCA. 
Class decomposition works by breaking down existing/predetermined classes. 
It is a form of equalising imbalanced classes.

Agglomerative Clustering following the rule of thumb: It was used on HOG’s and PCA features. 
It recursively merges the pair of clusters that minimally increases a given linkage distance. 
Linkage distance used was ‘ward’ which minimizes the variance of the clusters being merged. 
Rule of thumb is a simple way and follows the equation to find the number of clusters to be 
implemented for each class/target and follows the equation k = sqrt(n/2), where n is the number of objects.



### 3.1 FEATURE EXTRACTION --1-- HOG feature extraction

```python

#### Obtaining the HOG gradients of an image
from skimage import feature
class HOG:
    def __init__(self, orientations = 9, pixelsPerCell = (8, 8),
        cellsPerBlock = (2, 2), transform = False):
        self.orienations = orientations
        self.pixelsPerCell = pixelsPerCell
        self.cellsPerBlock = cellsPerBlock
        self.transform = transform
    def describe(self, image):
        hist = feature.hog(image, orientations = self.orienations,
            pixels_per_cell = self.pixelsPerCell,
            cells_per_block = self.cellsPerBlock,
            transform_sqrt = self.transform,
            block_norm = 'L2-Hys')
        return hist

HOG_Hist_data = []
#repo = []
for img in data_Bin:
    img = cv2.resize(img, (100,100))
    #img = cv2.imread(img)
    hog = HOG(orientations = 9, pixelsPerCell = (8, 8), cellsPerBlock = (2, 2), transform = True)
    hist = hog.describe(img)
    print('HOG Features for the image:')
    print(hist)
    print('Size of the HOG Features for the image:', hist.shape)
    HOG_Hist_data.append(hist)
```

### 3.4 FEATURE EXTRACTION --2-- PCA feature extraction

```python
# IMPORT
%matplotlib inline
#from torchvision import datasets
import PIL
from skimage.feature import local_binary_pattern, greycomatrix, greycoprops
from skimage.filters import gabor
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
#### duplication of data to be used for PCA feature extraction
imgData2 = imgData.copy()
trainDset = imgData2
print('No. of samples in train set: '+str(len(data_Bin)))
```

```python
#### Generating training data by extracting features from all images
import numpy as np
np.random.seed(100)
label = []
featLength = 2+5+2
trainFeats = np.zeros((len(trainDset),featLength)) #Feature vector of each image is of size 1x1030
for tr in range(len(trainDset)):
    print(str(tr+1)+'/'+str(len(trainDset)))
    
    image = trainDset[tr][0] #One image at a time
    #img_gray = img.convert('L') #Converting to grayscale
    image = cv2.resize(image, (100,100))
    #img_arr = np.array(img_gray.getdata()).reshape(img.size[1],img.size[0]) #Converting to array
    # LBP
    feat_lbp = local_binary_pattern(image,8,1,'uniform')
    #feat_lbp = local_binary_pattern(image,5,2,'uniform').reshape(image.size[0]*image.size[1])
    lbp_hist,_ = np.histogram(feat_lbp,8)
    lbp_hist = np.array(lbp_hist,dtype=float)
    lbp_prob = np.divide(lbp_hist,np.sum(lbp_hist))
    lbp_energy = np.nansum(lbp_prob**2)
    lbp_entropy = -np.nansum(np.multiply(lbp_prob,np.log2(lbp_prob)))   
    # GLCM
    gCoMat = greycomatrix(image, [2], [0],256,symmetric=True, normed=True)
    contrast = greycoprops(gCoMat, prop='contrast')
    dissimilarity = greycoprops(gCoMat, prop='dissimilarity')
    homogeneity = greycoprops(gCoMat, prop='homogeneity')    
    energy = greycoprops(gCoMat, prop='energy')
    correlation = greycoprops(gCoMat, prop='correlation')    
    feat_glcm = np.array([contrast[0][0],dissimilarity[0][0],homogeneity[0][0],energy[0][0],correlation[0][0]])
    # Gabor filter
    gaborFilt_real,gaborFilt_imag = gabor(image,frequency=0.6)
    gaborFilt = (gaborFilt_real**2+gaborFilt_imag**2)//2
    gabor_hist,_ = np.histogram(gaborFilt,8)
    gabor_hist = np.array(gabor_hist,dtype=float)
    gabor_prob = np.divide(gabor_hist,np.sum(gabor_hist))
    gabor_energy = np.nansum(gabor_prob**2)
    gabor_entropy = -np.nansum(np.multiply(gabor_prob,np.log2(gabor_prob)))
    # Concatenating features(2+5+2)    
    concat_feat = np.concatenate(([lbp_energy,lbp_entropy],feat_glcm,[gabor_energy,gabor_entropy]),axis=0)
    trainFeats[tr,:] = concat_feat #Stacking features vectors for each image
    # Class label
    label.append(trainDset[tr][1])
trainLabel = np.array(label) #Conversion from list to array
```

```python
#### Normalizing the train features to the range [0,1]
trMaxs = np.amax(trainFeats,axis=0) #Finding maximum along each column
trMins = np.amin(trainFeats,axis=0) #Finding maximum along each column
trMaxs_rep = np.tile(trMaxs,(2432,1)) #Repeating the maximum value along the rows
trMins_rep = np.tile(trMins,(2432,1)) #Repeating the minimum value along the rows
#np.divide(a[0],a[1], out=out, where=a[1]!=0)
trainFeatsNorm = np.divide(trainFeats-trMins_rep,trMaxs_rep,where=trMaxs_rep!=0) #Element-wise division
```

## 1. Introduction


#### Machine learning algorithms:
    
Support Vector Machine (SVM): a supervised machine learning algorithm which can be used for both 
classification or regression challenges. Here it is used for multiclass classification. In the SVM algorithm, 
we plot each data item as a point in n-dimensional space, with n being the number of features, 
with the value of each feature being the value of a particular coordinate. Then, classification 
is performed by finding the hyper-plane that differentiates the classes. Support Vectors are 
simply the co-ordinates of individual observation. For multiclass classification a one-versus-all 
approach is used for the creation of the hyper-plane for each class.

Random Forest: a versatile machine learning method capable of performing both regression and classification tasks. 
It is a type of ensemble learning method, where a group of weak models combine to form a powerful model. 
In Random Forest, multiple classification trees are grown. To classify a new object based on attributes, 
each tree gives a classification and the tree “votes” for that class. The forest chooses the classification having 
the most votes over all the trees in the forest.

Decision Trees:  Decision tree classifier is a systematic approach for multiclass classification. 
It poses a set of questions to the dataset (related to its attributes/features). The decision tree 
classification algorithm can be visualized on a binary tree. On the root and each of the internal nodes, 
a question is posed and the data on that node is further split into separate records that have different 
characteristics. The leaves of the tree refer to the classes in which the dataset is split. 

#### Parameter setting 

SVM, Random Forest, Desicion Trees:

For SVM gamma parameter was set to auto and all rest to default. 
For random forest n-estimators which is the number of trees in the forest was set to 20 
and the rest of the parameters were set to default. 
For Decision Trees criterion was set to “Gini Index” and max depth to 20, 40 and 70 according to the number of classes. When more classes existed in the dataset, bigger was the max depth,
referring to the maximum classes a branch could reach. It was not set to 39, 65 or 195 the number of the classes in the datasets, to avoid overfitting.

#### Neural Networks:
    
Artificial Neural Networks or ANN: A single perceptron (or neuron) can be imagined as a Logistic Regression. 
So ANN, is a group of multiple perceptrons/ neurons at each layer. ANN is capable of learning any nonlinear function.
ANNs have the capacity to learn weights that map any input to the output. One of the main reasons behind universal 
approximation is the activation function. Activation functions introduce nonlinear properties to the network.
This helps the network learn any complex relationship between input and output. One common problem in all these neural
networks is the Vanishing and Exploding Gradient. This problem is associated with the backpropagation algorithm.
The weights of a neural network are updated through this backpropagation algorithm by finding the gradients.

CNN: The building blocks of CNNs are filters a.k.a. kernels. Kernels are used to extract the relevant 
features from the input using the convolution operation. Convolving an image with filters results in a feature map. 
CNN learns the filters automatically without mentioning it explicitly. These filters help in extracting the right 
and relevant features from the input data. CNN captures the spatial features from an image. Spatial features refer 
to the arrangement of pixels and the relationship between them in an image. They help us in identifying the object 
accurately, the location of an object, as well as its relation with other objects in an image.

#### Parameter setting NN, CNN:
    
For both NN and CNN multiple activation functions, architectures and parameters were used. 
By obtaining results the best case was ultimately used. The activation functions tested were: 
“tanh”, “relu” and “sigmoid”, implemented for all architectures (different number of layers 
with different numbers of neurons). 

For NN the optimizer used was “SGD”, stochastic gradient descent, while for CNN “SGD” and “Adam”. 

For both NN and CNN the loss function was “sparse categorical crossentropy” which takes care of the multiclass
problem by automatically performing one hot encoding, meaning that there is a binary feature for each class value 
and the predictions must have predicted probabilities for each of the classes. The cross-entropy is then 
summed across each binary feature and averaged across all examples in the dataset.

Cross validation is supposed to take random folds of the dataset and use them as training set then applying the trained model on the randomly selected for this instance test set. The number of folds was chosen to be 5. The same procedure was followed for random forest and Decision tree machine learning algorithm.



#### Metrics chosen:

Accuracy: the ratio of correctly predicted observation to the total observations.

Recall: the ratio of correctly predicted positive observations to the all observations in actual class 

Precision: the ratio of correctly predicted positive observations to the total predicted positive observations.
F1-score: the weighted average of Precision and Recall. Therefore, this score takes both false positives and false negatives into account.

Matthews correlation coefficient (MCC)/phi-coefficient (φ): The Matthews correlation coefficient is used in machine learning as a measure of the quality of binary and multiclass classifications. It takes into account true and false positives and negatives and is generally regarded as a balanced measure which can be used even if the classes are of very different sizes. The MCC is in essence a correlation coefficient value between -1 and +1. A coefficient of +1 represents a perfect prediction, 0 an average random prediction and -1 an inverse prediction. The statistic is also known as the phi coefficient.

Cohen’s kappa coefficient/kappa score: measures the degree of agreement between the true values and the predicted values, which we use as the classifier’s performance. 1 means perfect agreement and 0 is chance agreement.

For NN the same augmented datasets were used with the same training/test set split. This way NN can be compared regarding all metrics with previous models. Metric of loss is used too. CNN though is applied on the augmented image dataset which is also imbalanced. Therefore, we cannot compare CNN with the previous models even though the same way of trainset/test set split is applied. Metrics used for evaluating the model were the same as for NN.
